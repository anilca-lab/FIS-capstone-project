```python
areas_df = pd.read_excel('area_definitions_m2018.xlsx')
metro_df = areas_df.loc[areas_df['May 2018 MSA code '].isin([35620, 31080, 16980, 19100, 26420, 33100, 37980, 12060, 71650, 47900])]
```

```python
metro_df[metro_df['County code'] == 25]
```

```python
state_code_list = ['0' + str(code) if len(str(code)) == 1 else str(code) for code in metro_df['FIPS code']]
```

```python
county_code_list = ['00' + str(code) if len(str(code)) == 1 else '0' + str(code) if len(str(code)) == 2 else str(code) for code in metro_df['County code']]
```

```python
fips = [s + c for s, c in zip(state_code_list, county_code_list)]
```

```python
import masked_array as ma
```

```python
values = metro_df.jk.tolist()
endpts = list(np.mgrid[min(values):max(values):5j])
fig = ff.create_choropleth(fips=fips, values=values, binning_endpoints=endpts, legend_title='Job vacancies by MSA', title='Labor Demand in the U.S.')
fig.layout.template = None
fig.show()
```
